```python
import pandas as pd
```

```python
names = pd.Series( data = ['Allen','Bob','Chris','Dave','Ed','Frank','Gus'])
gpas = pd.Series( data = [4.0, 3.0, 3.4, 2.8, 2.5, 3.8, 3.0])
years = pd.Series( data = ['So', 'Fr', 'Fr', 'Jr', 'Sr', 'Sr', 'Fr'])
series_dict = { 'Name':  names, 'GPA': gpas, 'Year' : years }  # dict of Series, keys are the series names
students = pd.DataFrame( series_dict )
students
```

```python
# filter columns only
students[ ['GPA', 'Year'] ]
```

```python
# a series of boolean values matching the condition
students['GPA'] > 3
```

```python
# filtering on the boolean condition
students[ students['GPA'] > 3 ]
```

```python
# combining both filters in the dataframe
students[ students['GPA'] > 3 ][ ['Name', 'GPA'] ]
```

```python
# printing dataframes with IPython.display
from IPython.display import display

print("Names and GPA's")
display(students[ ['Name', 'GPA']])
print("Juniors")
display(students[ students['Year'] =='Jr' ])
```

```python
# building an interactive program
year = input("Enter Class Year [Fr, So, Jr or Sr]: ")
print(f"{year} Students")
display(students[ students['Year'] == year ] )
```

```python
# you can assign to other variables
frosh = students[ students['Year'] == 'Fr' ]
display(frosh)
frosh_names = frosh[ ['Name']]
display(frosh_names)
```
