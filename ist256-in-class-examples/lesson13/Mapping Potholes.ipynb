```python
import folium
import pandas as pd
SYR = (43.0481, -76.1474)
syrmap = folium.Map(location=SYR, zoom_start=14)
potholes = pd.read_csv('https://cityofsyracuse.github.io/RoadsChallenge/data/potholes.csv')
for row in potholes.sample(200).to_records():
    coords = (row['Longitude'], row['Latitude'])
    pin= f"{row['strLocation'].title()} {row['dtTime']}"
    marker = folium.Circle(location = coords, radius= 10, popup=pin, color="#FF8600", fill_color="#FF8600")
    syrmap.add_child(marker)
    
syrmap
```
