```python
import IPython.display as ipd
from ipywebrtc import CameraStream, ImageRecorder
```

```python
my_web_cam = CameraStream(constraints= {'facing_mode': 'user', 'audio': False, 'video': { 'width': 640, 'height': 480 } })
ipd.display(my_web_cam)

image_recorder = ImageRecorder(stream=my_web_cam)
ipd.display(image_recorder)
```

```python
# let's do some image processing.
import PIL.Image
import PIL.ImageFilter
import io
```

```python
im = PIL.Image.open(io.BytesIO(image_recorder.image.value))
image_contour = im.filter(PIL.ImageFilter.CONTOUR)
ipd.display(image_contour)
```

```python
# save them to files
image_contour.save('fudge-contour.png')
image_recorder.save('fudge.png')
```
